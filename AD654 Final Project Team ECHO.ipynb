Part2: Segmentation and Targeting

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
segmentation = pd.read_csv("family_segments.csv")
```

```python
segmentation.head()
```

```python
#drop the irrelevant variable
segmentation = segmentation.drop("hhold_ID",axis = 1)
```

```python
#deal with the NA's
segmentation.isnull().sum()
```

```python
segmentation["oldest_child_age"] = segmentation.oldest_child_age.fillna(0)
```

```python
segmentation.describe()
```

```python
#check correlation
segmentation.corr()
```

```python
#data normalization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(segmentation)
data_normalized = scaler.transform(segmentation)
data_normalized = pd.DataFrame(data = data_normalized, index = segmentation.index, columns = segmentation.columns)
print(data_normalized.describe().round(2))
```

```python
sse = {}
for k in range (1,15):
    kmeans = KMeans(n_clusters=k, random_state = 654)
    kmeans.fit(data_normalized)
    sse[k] = kmeans.inertia_
plt.title("The Elbow Method")
plt.xlabel("k")
plt.ylabel("SSE")
sns.pointplot(x = list(sse.keys()),y = list(sse.values()))
```

```python
#start will 3 cluster, in where the distinct slope changes
kmeans = KMeans(n_clusters = 3, random_state = 654)
kmeans.fit (data_normalized)
cluster_labels = kmeans.labels_
```

```python
kmeans = data_normalized.assign(Cluster = cluster_labels)
kmeans.groupby(["Cluster"]).agg({
    "est_net_worth":"mean",
    "est_inc":"mean",
    "number_visits":"mean",
    "number_children":"mean",
    "oldest_child_age":"mean",
    "online_merch_avg":"mean",
    "est_annual_travel":"mean",
    "est_annual_leisure":"mean",
    "est_hhold_FICO":["mean","count"],
}).round(2)
```

```python
#try more clusters as no more differences between clusters
kmeans2 = KMeans(n_clusters = 5, random_state = 654)
kmeans2.fit (data_normalized)
cluster_labels = kmeans2.labels_
```

```python
kmeans2 = data_normalized.assign(Cluster = cluster_labels)
kmeans2.groupby(["Cluster"]).agg({
    "est_net_worth":"mean",
    "est_inc":"mean",
    "number_visits":"mean",
    "number_children":"mean",
    "oldest_child_age":"mean",
    "online_merch_avg":"mean",
    "est_annual_travel":"mean",
    "est_annual_leisure":"mean",
    "est_hhold_FICO":["mean","count"],
}).round(2)
```

```python
# try more clusters as the number of visit, aunnal leisure spending, net worth are not with sufficient distinctions.
kmeans3 = KMeans(n_clusters = 6, random_state = 654)
kmeans3.fit (data_normalized)
cluster_labels = kmeans3.labels_
```

```python
kmeans3 = data_normalized.assign(Cluster = cluster_labels)
kmeans3.groupby(["Cluster"]).agg({
    "est_net_worth":"mean",
    "est_inc":"mean",
    "number_visits":"mean",
    "number_children":"mean",
    "oldest_child_age":"mean",
    "online_merch_avg":"mean",
    "est_annual_travel":"mean",
    "est_annual_leisure":"mean",
    "est_hhold_FICO":["mean","count"],
}).round(2)
```

```python
pd.set_option('display.max.columns', None)
kmeans3.groupby("Cluster").describe()
```

```python
#visulization
df1 = kmeans3[["est_net_worth", "est_inc", "Cluster"]]
bar1 = pd.melt(df1, id_vars = "Cluster")
sns.catplot(x = 'Cluster', y='value', hue = 'variable', data=bar1, kind='bar')
```


This is a side-by-side bar plot that shows the distribution of the average household’s net worth and household’s annual income for each cluster. We can easily find that those two variables, net worth and annual income, for each cluster have a very similar pattern. Connected to the correlation table above, those two variables highly correlated each other.

```python
df2 = kmeans3[["est_annual_leisure", "est_annual_travel", "Cluster"]]
bar2 = pd.melt(df2, id_vars = "Cluster")
sns.catplot(x = 'Cluster', y='value', hue = 'variable', data=bar2, kind='bar')
```

There is another side-by-side bar plot which shows the distribution of annual travel spending and leisure spending of household for each cluster. Those variables also have a similar pattern for each cluster, that the cluster has higher travel spending also has higher leisure spending. Connect this plot with the one above, we find that one cluster that owns more money will spend more, and owns less will spend less.

```python
plot3 = sns.relplot(x = "number_visits",y = "est_annual_leisure", hue = "Cluster", data = kmeans3)
plot3.fig.set_figheight(8)
plot3.fig.set_figwidth(8)
```

This is a scatter plot shows the relationship between the household’s number of visits and its annual leisure spending for each cluster. For cluster 0, it visits most times but lowest leisure spending, compared with cluster 2, which has the least number of visits but the highest leisure spending. Within a similar number of visits, cluster 2 spends more than cluster 3 than cluster 1 and cluster 4 on leisure, so cluster 4 and 1 have the same level of lowest leisure spending, but cluster 4 visits fewer times than cluster 1. Cluster 5 does not have a clear pattern shown in this plot, with median leisure spending and median times of visit.

```python
plot4 = sns.relplot(x = "number_visits",y = "online_merch_avg", hue = "Cluster", size = "est_inc", data = kmeans3)
plot4.fig.set_figheight(8)
plot4.fig.set_figwidth(8)
```

This plot shows the relationship between visit times and average annual spending at online stores for each cluster at different income levels. For cluster 0, which has the highest visit times per year, its online expenditure depends on the income, i.e., higher-income households are willing to spend more online. Cluster 1 and 4 both have low online spending and low visit times, but cluster 1 has higher spending and more visits than cluster 4. Those 2 clusters, and also cluster 3 have low income. Cluster 2 and 3 have the same level of high online spending, but cluster 3 has a higher income level, and tends to have more visit times. ```python
vis = kmeans3.groupby("Cluster")[["number_children","online_merch_avg","est_hhold_FICO"]].mean()
sns.relplot(x = "number_children",y = "online_merch_avg", hue = "Cluster", size = "est_hhold_FICO", data = vis)
```

This scatter plot uses the mean value for each cluster to state the relationship between credit score and annual leisure spending. Cluster 4 has very low leisure spending and the lowest credit score, compared with cluster 1, which has the same low leisure spending, but higher credit score. Cluster 2 has very high spending on leisure per year, but a low credit score, whereas cluster 5 spends little on leisure, but has a very high credit score. This is surprising that families with higher credit scores would have more disposable money in hand and would have higher willingness to travel or spend on leisure.

Cluster 0 - “most-visit”: it contains families who have the most visit times per year. So the theme park can improve their service or on-site facilities or recreation to attract families in this cluster, some families have high income and are willing to pay more for the new attractions.

Cluster 1 - “big-family”: This cluster contains families with the most numbers of children and low income. Though they have low spending on travel and leisure, and low online purchases because of the budget, they still have some visits per year and not bad credit scores. So the theme park can offer them some group discounts or “spend more and get more back” deals to keep those customers.

Cluster 2 - “high-income”: This is a group of high-income families, and they also spend more on travel, leisure and online purchases. So we can have some fancy items or renew facilities to attract them, they will be willing to pay a higher price for anything they really interested in.

Cluster 3 - “online-buyer”: Families in this cluster have good income and there are also have big families. Compared with people will the same consumption levels, they pay more on online purchases. We can send them more coupon emails or online ads to incentive them to have more online purchases.

Cluster 4 - “poor”: This is a group with low income and low spending. They do not have big families and also low credit scores. In this sense, they are not in a good financial situation. So they are likely to see more offers, huge discounts, or some special events with promotions.

Cluster 5 - “Dink”: This is a cluster of families with no kids and the highest credit scores. They seldomly worried about the price, though, their other preferences may vary, as they may be young new couples or empty-nester couples, busy-workers or retired couples. ```python
def first_day(entry):
    """
    Returns the first instance of the period, regardless of sampling rate.
    """
    if len(entry):  # handles the case of missing data
        return entry[0]
```

```python
cruise1 = pd.get_dummies(cruise, drop_first=False, columns=['cabin_type', 'season_of_year', 'cruise_theme', 'embark_port'])
```

```python
cruise1.head()
```

```python
# show data type
cruise1.info()
```

```python
cruise1.columns
```

```python
x=cruise1[['discount_original', 'coupon_received', 'est_inc', 'ticket_spend',
       'incidental_spending', 'park_package', 'res_to_port', 'number_children',
       'oldest_child_age', 'dis_plus_sub', 
       'cabin_type_Concierge', 'cabin_type_Inside', 'cabin_type_Oceanview',
       'cabin_type_Verandah', 'season_of_year_Fall', 'season_of_year_Spring',
       'season_of_year_Summer', 'season_of_year_Winter', 'cruise_theme_Marvel',
       'cruise_theme_No_theme', 'cruise_theme_Princess/Frozen',
       'cruise_theme_Star Wars', 'cruise_theme_prin_fro',
       'cruise_theme_starwars', 'embark_port_Barcelona',
       'embark_port_Copenhagen', 'embark_port_New York',
       'embark_port_Port Canaveral', 'embark_port_San Diego',
       'embark_port_Vancouver']]
y=cruise1['return_1065']
```

```python
# We set 60% data for training set, 40% data for test set
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(x, y, test_size=0.4,random_state=654)
```

```python
# Standlize our training set data
from sklearn.preprocessing import StandardScaler
scaler1 = StandardScaler()
scaler1.fit(x_train1)
x_train1_normalized = scaler1.transform(x_train1)
x_train1_normalized = pd.DataFrame(data=x_train1_normalized, index=x_train1.index, columns=x_train1.columns)
scaler2 = StandardScaler()
scaler2.fit(x_test1)
x_test1_normalized = scaler2.transform(x_test1)
x_test1_normalized = pd.DataFrame(data=x_test1_normalized, index=x_test1.index, columns=x_test1.columns)
```

```python
# Use logistic regression to create a model
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train1_normalized, y_train1)
```

```python
pd.DataFrame(data=logmodel.coef_.transpose(),index=x_train1_normalized.columns,columns=['Coef'])
```

```python
# The accuracy of test set
prediction1 = logmodel.predict(x_test1_normalized)
from sklearn.metrics import accuracy_score
accuracy_score(y_test1, prediction1)
```

```python
# The accuracy of training set
prediction2 = logmodel.predict(x_train1_normalized)
accuracy_score(y_train1, prediction2)
```

Part6: AB test

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import chisquare
```

```python
popcorn = pd.read_csv("popcorn_buckets.csv")
popcorn
```

```python
import scipy.stats as stats
(t1, p1)= stats.ttest_ind(popcorn.loc[popcorn['bucket']=='Cinderella','Q1_sales_USD'].values,
                          popcorn.loc[popcorn['bucket']=='Mermaid','Q1_sales_USD'].values, equal_var=False)
print("The test statistic is: ", round(t1,5))
print("The p-value is: ", round(p1,5))
```

```python
import scipy.stats as stats
(t2, p2)= stats.ttest_ind(popcorn.loc[popcorn['bucket']=='MickeyEars','Q1_sales_USD'].values,
                          popcorn.loc[popcorn['bucket']=='MickeySuit','Q1_sales_USD'].values, equal_var=False)
print("The test statistic is: ", round(t2,5))
print("The p-value is: ", round(p2,5))
```

```python
import scipy.stats as stats
(t3, p3)= stats.ttest_ind(popcorn.loc[popcorn['bucket']=='RedWhite','Q1_sales_USD'].values,
                          popcorn.loc[popcorn['bucket']=='Star_Wars','Q1_sales_USD'].values, equal_var=False)
print("The test statistic is: ", round(t3,5))
print("The p-value is: ", round(p3,10))
```

0.018<0.05, so we reject null hypothesis, we choose Mermaid.

0.063>0.05, so we fail to reject null hypothesis, we choose MickeyEars.

7.00417e-05<0.05, so we reject null hypothesis, we choose Star_wars.
